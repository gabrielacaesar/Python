```python
def buscar_deputados():
  deputados = []
  for n in range(1,7):
    url = f'https://dadosabertos.camara.leg.br/api/v2/deputados?itens=100&pagina={n}'
    r = requests.get(url)
    for linha in r.json()["dados"]:
      deputados.append(linha)
  return deputados

deputados = buscar_deputados()
```

```python
chrome_path = "c:\\Users\\abraji\\downloads\\chromedriver_win32\\chromedriver.exe"

def encontrar_no_twitter(nome):
    nome = nome.replace(" ", "+")
    busca = f'https://www.google.com.br/search?q=twitter+{nome}+deputado'
    browser = webdriver.Chrome(chrome_path)
    browser.get(busca)
    html = browser.page_source
    soup = bs(html, "html.parser")
    browser.close()
    links = []
    for link in soup.find_all("a"):
        try:
            href = link["href"]
            if "twitter.com" in href:
                if "status" in href:
                    return "Não encontrado"
                else:
                    return href
        except KeyError:
            continue
    return links[0]
```

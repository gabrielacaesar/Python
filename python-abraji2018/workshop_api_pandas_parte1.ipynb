```
def buscar_deputados():
  deputados = []
  for n in range(1,7):
    url = f'https://dadosabertos.camara.leg.br/api/v2/deputados?itens=100&pagina={n}'
    r = requests.get(url)
    for linha in r.json()["dados"]:
      deputados.append(linha)
  return deputados
```

```
deputados = buscar_deputados()
```
